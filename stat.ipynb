```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from math import pi
```

```python
# Set data
df = pd.DataFrame({
'stats': ['Power', 'Balance', 'Coordination', 'Endurance', 'Control'],
'23 OCT': [3, 3.2, 2.8, 2.9, 3.1],
'23 DEC': [3.3, 3.3, 3.1, 3.2, 3.3],
'24 FEB': [3.9, 3.7, 3.6, 3.5, 3.8],
})
# number of variable
categories=df['stats']
#print(categories)
N = len(categories)

# angle of each axis
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# plot inint
plt.figure(figsize=(11, 14))
ax = plt.subplot(111, polar=True)
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
labels = []
for label,i in zip(ax.get_xticklabels(),range(0,len(angles))):

    angle_rad=angles[i]
    if angle_rad <= 2*pi/3:
        ha= 'left'
        va= "top"

    elif 2*pi/3 < angle_rad <= 4*pi/3:
        ha= 'center'
        va= "top"

    elif 4*pi/3 < angle_rad <= (2*pi):
        ha= 'right'
        va= "bottom"  

    else:
        ha= 'right'
        va= "bottom"

    label.set_rotation(angles[i])
    label.set_verticalalignment(va)
    label.set_horizontalalignment(ha)
    label.set_fontsize(14)
    pos = label.get_position()
    label.set_position((pos[0], pos[1]-0.05))
    labels.append(label)

# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([2,2.5, 3, 3.5, 4], ["V2","","V3","", "V4"], alpha=0.5)
plt.ylim(2, 4.0)
ytick_colors = ['green', 'black', 'blue', 'black', 'red']

# Yticklabel styles
for label, i in zip(ax.get_yticklabels(), range(0, len(angles))):
    label.set_fontsize(14)
    #label.set_fontweight('bold')
    label.set_color('black')
    label.set_verticalalignment('top')
    label.set_horizontalalignment('center')

for label, color in zip(ax.get_yticklabels(), ytick_colors):
    label.set_color(color)
    label.set_fontsize(17)
    label.set_fontweight('bold')
    label.set_position((-0.8, -1))

colors = ['black', 'mediumblue', 'cyan']

keys = df.keys()
for i in range(1, len(keys)):
    values=df[keys[i]].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=keys[i], color=colors[i-1])
    ax.fill(angles, values, color=colors[i-1], alpha=0.2)

ax.spines['polar'].set_visible(False)
ax.legend(loc='upper left', bbox_to_anchor=(-0.15, 1.05), fontsize=15)#, prop = {'weight':'bold'})
ax.set_title("Climbing Skill Status", size=20, color='black', va='top', ha='center', fontweight='bold', pad=100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define categories and elements
categories = ['Strength', 'Technique', 'Endurance', 'Coordination', 'Control']
elements = ['Power', 'Grip', 'Precision', 'Balance', 'Flexibility']

# Define data for each element (example values)
data = {
    'Power': [8, 7, 6, 5, 7],
    'Grip': [7, 9, 6, 4, 8],
    'Precision': [6, 8, 7, 6, 7],
    'Balance': [5, 6, 7, 8, 6],
    'Flexibility': [7, 6, 5, 6, 7]
}

# Calculate angles for radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()

# Make the plot close to a circle
data = np.concatenate((data, [data[0]]))
angles += angles[:1]

# Plot radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for i in range(len(elements)):
    ax.plot(angles, data[elements[i]], label=elements[i])
ax.fill(angles, data[elements[0]], alpha=0.1)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Add category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Show the plot
plt.show()
```
